7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

df=pd.DataFrame(df)

# Encode the 'role' column
label_encoder = LabelEncoder()
df['role'] = label_encoder.fit_transform(df['role'])

# Separate features and target variable
X = df.drop(columns=['hero_name', 'role', 'ban_rate', 'pick_rate', 'release_year', 'win_rate'])
y = df['role']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create a Decision Tree Classifier
classifier = DecisionTreeClassifier(random_state=10)

# Train the model
classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Character specifications to predict the role
new_character = pd.DataFrame({
    "defense_overall": [6.0],
    "offense_overall": [7.0],
    "skill_effect_overall": [4.0],
    "difficulty_overall": [3.0],
    "movement_spd": [250],
    "magic_defense": [9],
    "mana": [380],
    "hp_regen": [37],
    "physical_atk": [119],
    "physical_defense": [24],
    "hp": [2689],
    "attack_speed": [0.75],
    "mana_regen": [16.0]
})

# Predict the role for the given character
predicted_role = classifier.predict(new_character)

# Decode the predicted role back to original labels
predicted_role_label = label_encoder.inverse_transform(predicted_role)

# Print the predicted role
print("Predicted role for the character:", predicted_role_label[0])
```

8. Why is CSV one of the most commonly used file formats in the field of data science? <br>

# 8
CSV (Comma-Separated Values) is a widely used file format in data science, valued for its simplicity and readability as it consists of plain text that can be easily edited and understood. The format is universally compatible across various platforms and programming languages, with built-in libraries available for reading and writing in languages like Python, R, and Java, making it easy to integrate with tools like Excel and Google Sheets. CSV provides a standardized way to represent tabular data, facilitating efficient data interchange in ETL processes. It is typically more storage-efficient than other formats, resulting in faster read and write times due to minimal overhead. Additionally, CSV accommodates various data types and allows for customizable delimiters, enhancing its flexibility. The format also supports efficient batch processing and manipulation through libraries like Pandas, and is commonly used as input for machine learning models, particularly in feature engineering. Overall, the combination of these attributes makes CSV an essential component of data science workflows.
